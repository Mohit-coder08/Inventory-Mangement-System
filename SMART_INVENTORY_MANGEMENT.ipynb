```python
class Authentication:
    def __init__(self):
        self.username = "admin"
        self.password = "admin123"

    def login(self):
        attempts = 3

        while attempts > 0:
            print("\n===== SMART INVENTORY SYSTEM LOGIN =====")
            username = input("Enter Username: ")
            password = input("Enter Password: ")

            if username == self.username and password == self.password:
                print("\nLogin Successful!\n")
                return True
            else:
                attempts -= 1
                print(f"Invalid Credentials! Attempts Left: {attempts}")

        print("Access Denied!")
        return False
```

```python
class ProductManagement:
    def __init__(self):
        self.products = {}

    def add_product(self):
        pid = input("Product ID: ")

        if pid in self.products:
            print("Product already exists!")
            return

        self.products[pid] = {
            "name": input("Product Name: "),
            "category": input("Category: "),
            "cost": float(input("Cost Price: ")),
            "price": float(input("Selling Price: ")),
            "quantity": int(input("Quantity: ")),
            "supplier": input("Supplier: ")
        }

        print("Product Added Successfully!")

    def view_products(self):
        if not self.products:
            print("No Products Available.")
            return

        print("\nID\tName\tCategory\tQty\tPrice")
        for pid, p in self.products.items():
            print(f"{pid}\t{p['name']}\t{p['category']}\t{p['quantity']}\t{p['price']}")

    def search_product(self):
        pid = input("Enter Product ID: ")

        if pid in self.products:
            print(self.products[pid])
        else:
            print("Product Not Found!")

    def update_product(self):
        pid = input("Enter Product ID: ")

        if pid not in self.products:
            print("Product Not Found!")
            return

        self.products[pid]["price"] = float(input("New Selling Price: "))
        self.products[pid]["quantity"] = int(input("New Quantity: "))

        print("Product Updated Successfully!")

    def delete_product(self):
        pid = input("Enter Product ID: ")

        if pid in self.products:
            del self.products[pid]
            print("Product Deleted Successfully!")
        else:
            print("Product Not Found!")
```

```python
class InventoryManagement:
    def __init__(self, products):
        self.products = products

    def display_stock(self):
        print("\nCurrent Stock")
        for pid, product in self.products.items():
            print(product["name"], ":", product["quantity"])

    def restock(self):
        pid = input("Enter Product ID: ")

        if pid in self.products:
            qty = int(input("Add Quantity: "))
            self.products[pid]["quantity"] += qty
            print("Stock Updated!")
        else:
            print("Product Not Found!")

    def low_stock(self):
        print("\nLow Stock Products")

        for product in self.products.values():
            if product["quantity"] < 5:
                print(product["name"], "-", product["quantity"])
```

```python
class SupplierManagement:
    def __init__(self):
        self.suppliers = {}

    def add_supplier(self):
        sid = input("Supplier ID: ")

        if sid in self.suppliers:
            print("Supplier already exists!")
            return

        self.suppliers[sid] = {
            "company": input("Company Name: "),
            "phone": input("Phone: "),
            "email": input("Email: "),
            "address": input("Address: ")
        }

        print("Supplier Added Successfully!")

    def view_suppliers(self):
        if not self.suppliers:
            print("No Suppliers Available.")
            return

        print("\nSupplier List")
        for sid, s in self.suppliers.items():
            print(f"{sid} | {s['company']} | {s['phone']}")

    def search_supplier(self):
        sid = input("Enter Supplier ID: ")

        if sid in self.suppliers:
            print(self.suppliers[sid])
        else:
            print("Supplier Not Found!")

    def delete_supplier(self):
        sid = input("Enter Supplier ID: ")

        if sid in self.suppliers:
            del self.suppliers[sid]
            print("Supplier Deleted Successfully!")
        else:
            print("Supplier Not Found!")
```

```python
class CustomerManagement:
    def __init__(self):
        self.customers = {}

    def add_customer(self):
        cid = input("Customer ID: ")

        if cid in self.customers:
            print("Customer already exists!")
            return

        self.customers[cid] = {
            "name": input("Customer Name: "),
            "phone": input("Phone: "),
            "email": input("Email: "),
            "history": []
        }

        print("Customer Added Successfully!")

    def search_customer(self):
        cid = input("Enter Customer ID: ")

        if cid in self.customers:
            print(self.customers[cid])
        else:
            print("Customer Not Found!")

    def purchase_history(self):
        cid = input("Enter Customer ID: ")

        if cid in self.customers:
            print("\nPurchase History")
            history = self.customers[cid]["history"]

            if history:
                for item in history:
                    print(item)
            else:
                print("No Purchases Yet.")
        else:
            print("Customer Not Found!")
```

```python
class SalesManagement:
    def __init__(self, products, customers):
        self.products = products
        self.customers = customers
        self.sales = []

    def sell_product(self):
        cid = input("Customer ID: ")

        if cid not in self.customers:
            print("Customer Not Found!")
            return

        pid = input("Product ID: ")

        if pid not in self.products:
            print("Product Not Found!")
            return

        qty = int(input("Quantity: "))

        if qty > self.products[pid]["quantity"]:
            print("Insufficient Stock!")
            return

        self.products[pid]["quantity"] -= qty

        total = qty * self.products[pid]["price"]

        sale = {
            "customer": self.customers[cid]["name"],
            "product": self.products[pid]["name"],
            "quantity": qty,
            "total": total
        }

        self.sales.append(sale)
        self.customers[cid]["history"].append(sale)

        print("\nSale Successful!")
        print("Total Amount: ₹", total)

    def view_sales(self):
        if not self.sales:
            print("No Sales Yet.")
            return

        print("\nSales Report")
        for sale in self.sales:
            print(sale)
```

```python
class PurchaseManagement:
    def __init__(self, products):
        self.products = products
        self.purchases = []

    def purchase_stock(self):
        pid = input("Enter Product ID: ")

        if pid not in self.products:
            print("Product Not Found!")
            return

        qty = int(input("Purchase Quantity: "))
        cost = float(input("Cost Price: "))

        self.products[pid]["quantity"] += qty
        self.products[pid]["cost"] = cost

        purchase = {
            "product": self.products[pid]["name"],
            "quantity": qty,
            "cost": cost
        }

        self.purchases.append(purchase)

        print("Stock Purchased Successfully!")

    def view_purchases(self):
        if not self.purchases:
            print("No Purchase Records.")
            return

        print("\nPurchase History")
        for purchase in self.purchases:
            print(purchase)
```

```python
class BillingSystem:
    def generate_bill(self, product_name, price, quantity):
        subtotal = price * quantity

        discount = 0

        if subtotal >= 5000:
            discount = subtotal * 0.10

        gst = (subtotal - discount) * 0.18
        total = subtotal - discount + gst

        print("\n========== INVOICE ==========")
        print("Product :", product_name)
        print("Price   :", price)
        print("Quantity:", quantity)
        print("-----------------------------")
        print("Subtotal :", subtotal)
        print("Discount :", discount)
        print("GST (18%):", round(gst, 2))
        print("-----------------------------")
        print("Grand Total :", round(total, 2))
        print("=============================")

        return total
```

```python
class Reports:
    def __init__(self, products, sales):
        self.products = products
        self.sales = sales

    def inventory_report(self):
        print("\n===== INVENTORY REPORT =====")

        total_value = 0

        for pid, product in self.products.items():
            value = product["quantity"] * product["cost"]
            total_value += value

            print(
                f"{product['name']} | Stock: {product['quantity']} | Value: ₹{value}"
            )

        print("\nTotal Inventory Value: ₹", total_value)

    def sales_report(self):
        print("\n===== SALES REPORT =====")

        total_sales = 0

        for sale in self.sales:
            print(sale)
            total_sales += sale["total"]

        print("\nTotal Sales: ₹", total_sales)

    def low_stock_report(self):
        print("\n===== LOW STOCK REPORT =====")

        found = False

        for product in self.products.values():
            if product["quantity"] < 5:
                found = True
                print(product["name"], "-", product["quantity"])

        if not found:
            print("No Low Stock Products.")
```
